````python
# Load Data
df = pd.read_csv('2021-2022 Football Player Stats.csv', sep = ';')
pd.set_option('display.max_columns', None)
df.head()
````

````python
df.columns.values
````

### Feature selection of raw DataFrame with Specific Stats

````python
### Key Stats
stats = df[['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', # Categories
            # Numeric Stats
            'Min', 'Shots', 
            'PasAss', # Pass leading to a shot
            'PasTotAtt', 'PasTotCmp', # Passing 
            'DriAtt', 'DriPast', 'CarPrgDist', 'Fld', # Driblling(Runs with ball)
            'Tkl+Int', 'BlkSh', 'Clr', 'Fls' ]]# Defense
stats.head()
````

````python
stats.shape
````

````python
# 2921 Players
# Decent sample size via minutes played 

min_played = stats['Min'].median()
print('Selecting Players with Minutes Played over:', min_played)
matches = round(stats['Min'].median()/90)
print('Players with at least:', matches, 'matches played')
stats = stats.loc[stats['Min'] >= min_played]
stats.shape
````

### Q1: Footballers in EFIGS by postion [Pos]

````python
# Segment by postion
stats['Pos'].unique()
````

````python
# Postion Modifiers
stats = stats.replace({'Pos' : {'DF' : 'DEF',
                                'MF' : 'MID',
                                'FW' : 'ATT',
                                'MFFW': 'MID',
                                'FWMF': 'ATT',
                                'DFMF': 'DEF',
                                'FWDF': 'ATT', 
                                'MFDF': 'MID',
                                'DFFW': 'DEF',
                                'GKMF': 'GK'}})
print('Postions:', stats['Pos'].unique())
````

````python
demo = stats[['Nation', 'Pos']]
demo.head()
````

````python
# Players within a nation
EFIGS_pp = stats.groupby(['Nation', 'Pos']).size().unstack().fillna(0) #.reset_index()
# Totals
EFIGS_pp['pp_Total'] = EFIGS_pp.sum(axis=1, numeric_only= True)
# Selecting top 15 nations
EFIGS_pp = EFIGS_pp.sort_values(by=['pp_Total'], ascending=False).head(15) #top15
EFIGS_pp = EFIGS_pp.drop(columns=['pp_Total']) #Dropping the counts
EFIGS_pp.head()
````

````python
EFIGS_pp.plot(kind='bar', stacked=True,
              figsize=(10,6)).legend(loc='upper right')

# Text Cues
plt.title('Player Counts for each Postion by Nation', 
         fontsize=16)
plt.xlabel("Nation", fontsize=12)
plt.xticks(rotation='horizontal') #rot=0
plt.ylabel("Player Counts", fontsize=12)

# Show
plt.show()
````

## Q2: Ages by Compeitition

````python
# Individual Age Boxplot by Competition

fig2 = stats.boxplot(column=['Age'], by=['Comp'],
             figsize=(6,8),
                     patch_artist=True,
                     boxprops=dict(facecolor='darkgreen', color='r'), #all boxes
                     medianprops= dict(color='whitesmoke', linewidth=2.5)) #median

# Text Cues
plt.title('Player Ages by Competition', 
         fontsize=16)
plt.xlabel("Competition", fontsize=12)
plt.xticks(rotation='horizontal') #rot=0
plt.ylabel("Age", fontsize=12)
plt.yticks(np.arange(15, 45, 5)) # frequency

# Show
plt.grid(False)
plt.show()
````

### Q3: Chance Creation - Players by shots and passes leading to shots per 90

````python
impact = stats[['Player', 'PasAss', 'Shots']]
impact
# Totals
impact['chances'] = impact['PasAss'] + impact['Shots']
impact = impact.sort_values(by='chances', ascending=False).head(15) #sorting
impact = impact.drop(columns=['chances']) #Dropping the counts
impact.head()
````

````python
impact_fig, ax = plt.subplots(figsize=(10, 6))


ax.scatter(x=impact['Shots'], y=impact['PasAss'])

players = impact['Player'].to_list()

# Text Cues
plt.title('Player Attacking Impact', 
         fontsize=16)
plt.xlabel("Shots per 90", fontsize=12)
plt.ylabel("Layoffs per 90", fontsize=12)


plt.show()
````
